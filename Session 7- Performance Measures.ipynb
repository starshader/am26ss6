# Performance Measures of Classification 

Let X1, X2, ....., Xn be the features.

 Z= b0 + b1 X1 + b2 X2 + .... + bn Xn
 
 f(Z) = e ^ Z / ( 1+ e^ Z) 
 
 

if 0< f(Z)< 0.5. ===> Class 0

if 0.5< f(Z) < 1 ===> Class 1

Two classes: Binary classification

Logistic regression is used for binary classification.

## Sourcing the data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Binary classification using LR for German Credit problem

```python
gc=pd.read_csv('https://online.stat.psu.edu/onlinecourses/sites/stat508/files/german_credit.csv')
```

```python
gc
```

```python
gc.info()
```

## Data preparation

```python
y= gc['Creditability']
```

```python
y.unique()
```

```python
# Only two classes- fit for Logistic regression
```

```python
X=gc.drop(['Creditability'],axis=1)
X.head()
```

```python
X.shape
```

```python
X.columns
```

```python
cat_cols=['Account Balance', 
       'Payment Status of Previous Credit', 'Purpose', 
       'Value Savings/Stocks', 'Length of current employment',
       'Instalment per cent', 'Sex & Marital Status', 'Guarantors',
       'Duration in Current address', 'Most valuable available asset',
       'Concurrent Credits', 'Type of apartment',
       'No of Credits at this Bank', 'Occupation', 'No of dependents',
       'Telephone', 'Foreign Worker']
```

```python
len(cat_cols)
```

```python
# Converting categorical features to numeric using 1-hot encoding

X_1= pd.get_dummies(X, columns=cat_cols)
X_1.shape
```

```python
# Adding constant 1
import statsmodels.api as sm
X_1= sm.add_constant(X_1)
```

```python
X_1.shape
```

```python
X_1.iloc[0:10,0:20]
```

```python
X_1.iloc[0:10,20:40]
```

```python
X_1.iloc[0:10,40:60]
```

```python
X_1.iloc[0:10,60:]
```

## Splitting to: Train and test

```python
from sklearn.model_selection import train_test_split
X_train_1,X_test_1,y_train_1,y_test_1=train_test_split(
     X_1,y, test_size=0.2,random_state=10)

X_train_1.shape,X_test_1.shape,y_train_1.shape,y_test_1.shape
```

## Building the model


```python
logR_1=sm.Logit(y_train_1,X_train_1)

# Training the model
logR_1=logR_1.fit()
```

## Diagnosing the model

```python
logR_1.summary2()
```

```python
signi_features_1=['Duration of Credit (month)','Credit Amount',
                 'Age (years)']
```

```python
X_2=X_1[signi_features_1]
```

```python
X_2
```

## Continue with the revised feature set X_2

```python
X_train_2,X_test_2,y_train_2,y_test_2=train_test_split(
     X_2,y, test_size=0.2,random_state=10)

X_train_2.shape,X_test_2.shape,y_train_2.shape,y_test_2.shape
```

```python
# Build the model

logR_2=sm.Logit(y_train_2,X_train_2)
logR_2=logR_2.fit()
logR_2.summary2()
```

```python
signi_features_2=['Duration of Credit (month)',
                 'Age (years)']
```

```python
X_3=X_2[signi_features_2]
```

```python
X_3
```

## Conitune with the revised feature set X_3

```python
X_train_3,X_test_3,y_train_3,y_test_3=train_test_split(
     X_3,y, test_size=0.2,random_state=10)

X_train_3.shape,X_test_3.shape,y_train_3.shape,y_test_3.shape
```

```python
logR_3=sm.Logit(y_train_3,X_train_3)
logR_3=logR_3.fit()
logR_3.summary2()
```

```python
logR_3.params
```

The final model:

Z = Duration of Credit (month) * -0.031934 +
    Age (years) * 0.044396
    
f(Z) = e^Z / (1+e^Z) ==> Logistic fn



## Predicting using the model

```python
y_pred= logR_3.predict(X_test_3)
y_pred
```

```python
y_test_3
```

```python
# Create a DF

pred_df=pd.DataFrame({'Actual Class':y_test_3,
                     'Predicted Prob': y_pred})
pred_df
```

```python
pred_df['Predicted Class']=pred_df['Predicted Prob'].map(lambda x: 1 if x>0.5 else 0)
```

```python
pred_df
```

## ROC Curve

- Receiver Operating Characteristics Curve

A curve plotted by taking:
  FPR - Along X- axis ( sensitivity)
  
  TPR - Along Y - axis ( 1- specificity)
  

```python
from sklearn.metrics import roc_curve

fpr,tpr,threshold=roc_curve(pred_df['Actual Class'],
                           pred_df['Predicted Prob'])

plt.plot(fpr,tpr);
```

```python
plt.plot(fpr,tpr);
plt.plot([0,1],[0,1],color='r')
plt.xlabel(' FPR')
plt.ylabel('TPR');
```

```python
# Area under the roc curve

from sklearn.metrics import roc_auc_score

score= roc_auc_score(pred_df['Actual Class'],
                           pred_df['Predicted Prob'])

print(' ROC-AUC-Score:',score)
```

## Finding optimal classification cut-off

- Instead of default 0.5



## Youden's index

Youden's index= Max( senstivity + specificity - 1)

              = Max( senstivity - (1-  specificity)
              
              = Max (FPR - TPR)

```python
# Create a DF

fpr_tpr=pd.DataFrame({'FPR':fpr,
                     'TPR':tpr,
                     'Threshold':threshold})
```

```python
fpr_tpr
```

```python
fpr_tpr['Diff']=fpr_tpr['TPR']-fpr_tpr['FPR']
fpr_tpr
```

```python
fpr_tpr.sort_values('Diff',ascending=False)
```

Remark:
    
As per Youdens index, find the optimum threshold based 
on the maximum Differece as the cut-off = 0.526841



## The new predicted class

```python
pred_df['New Predicted Class']=pred_df['Predicted Prob'].map(lambda x: 1 if x>0.526841 else 0)
```

```python
pred_df
```

```python
cm_new= confusion_matrix(pred_df['Actual Class'],
                        pred_df['New Predicted Class'])

print('The new confusion matrix:\n', cm_new)
```

```python
score_new=roc_auc_score(pred_df['Actual Class'],
                       pred_df['New Predicted Class'])
print('The new ROC-AUC-Score',score_new)
```
