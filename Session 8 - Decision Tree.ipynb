## Preprocessing the data

## Splitting the dataset: train and test

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y, test_size=0.2,
                                               random_state=1)

X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

## Building the model

```python
from sklearn.tree import DecisionTreeClassifier

dt_1=DecisionTreeClassifier()

dt_1=dt_1.fit(X_train,y_train)
```

```python
dt_1
```

## Prediction using the model

```python
y_pred=dt_1.predict(X_test)

y_pred
```

```python
y_test
```

## Checking the preformance of the model

```python
from sklearn.metrics import confusion_matrix,classification_report,roc_curve,roc_auc_score
```

```python
cm=confusion_matrix(y_test,y_pred)
sns.heatmap(cm,annot=True);
```

```python
report=classification_report(y_test,y_pred)
print(' The classification report:\n',report)
```

```python
fpr,tpr,_=roc_curve(y_test,y_pred)
plt.plot(fpr,tpr);
```

## Visualising the tree created

```python
from sklearn.tree import plot_tree

plt.figure(figsize=(15,15))
plot_tree(dt_1,filled=True);
```

## Improving the model

```python
dt_2=DecisionTreeClassifier(criterion='entropy',max_depth=4)

dt_2=dt_2.fit(X_train,y_train)
```

```python
y_pred_2=dt_2.predict(X_test)
```

```python
cm_2=confusion_matrix(y_test,y_pred_2)

sns.heatmap(cm_2,annot=True);
```

```python
plt.figure(figsize=(15,15))
plot_tree(dt_2,filled=True);
```
