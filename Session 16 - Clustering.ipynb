## Implement KMeans on this dataset

### K using Elbow method

```python
from sklearn.cluster import KMeans
SSD=[]

for k in range(1,25):
    kmeans=KMeans(n_clusters=k, random_state=10)
    kmeans.fit(X)
    SSD.append(kmeans.inertia_)
plt.plot(range(1,25),SSD);
# plt.xlim([0,6]);
```

The best value of k =4

### Silhouette method

```python
# Silhouette score

from sklearn.metrics import silhouette_score

SS=[]
for k in range(2,25):
    kmeans=KMeans(n_clusters=k, random_state=10)
    kmeans.fit(X)
    SS.append(silhouette_score(X,kmeans.predict(X)))
plt.plot(range(2,25),SS);
```

The best value of k =4

## Implementing the model using k=4

```python
k_best=KMeans(n_clusters=4, random_state=10)
k_best.fit(X)
clusters=k_best.predict(X)
print(' The cluser labels are:\n', clusters)
```

```python
# The cluster centroids

k_best.cluster_centers_
```

## Implementing KMeans on a real data set

```python
from sklearn.datasets import load_iris
iris=load_iris()
iris
```

```python
# Converting numpy array to pd DF

X=pd.DataFrame(iris['data'],columns=['SL','SW','PL','PW'])
X
```

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()

X_scaled=scaler.fit_transform(X)
X_scaled
```

```python
# Elbow method to find the value of k


SSD=[]

for k in range(1,25):
    kmeans=KMeans(n_clusters=k, random_state=10)
    kmeans.fit(X_scaled)
    SSD.append(kmeans.inertia_)
plt.plot(range(1,25),SSD)
plt.xlim([0,6])
```

The best value of k =3

```python
# Building the model using this value of k=3.

k_final=KMeans(n_clusters=3, random_state=10)
k_final.fit(X_scaled)
clusters=k_final.predict(X_scaled)
print(' The clusters formed are:\n',clusters)
```

```python
k_final.cluster_centers_
```
