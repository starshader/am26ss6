## Accessing the dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
oly=pd.read_csv('olympics.csv')
```

```python
oly
```

## Preprocessing


```python
X=oly['Year']
X
```

```python
y=oly['Time']
y
```

## Plotting 

```python
plt.figure(figsize=(10,8))
plt.scatter(X,y);
```

## Simple LR

```python
from sklearn.linear_model import LinearRegression

lr=LinearRegression()

lr.fit(X.to_numpy().reshape(-1,1),y)
```

```python
plt.figure(figsize=(10,8))
plt.scatter(X,y)
plt.plot(X,lr.predict(X.to_numpy().reshape(-1,1)),label='Lin Reg')
plt.legend();
```

## Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures

p2=PolynomialFeatures(degree=2)

Xp2=p2.fit_transform(X.to_numpy().reshape(-1,1))
Xp2
```

```python
X
```

```python
lr2=LinearRegression()

pr2=lr2.fit(Xp2,y)
```

```python
# Plotting

plt.figure(figsize=(10,8))
plt.scatter(X,y)
plt.plot(X,lr.predict(X.to_numpy().reshape(-1,1)),label='Lin Reg')

plt.plot(X,pr2.predict(Xp2),label='Poly Reg degree 2')

plt.legend();
```

## Polynomial regression degree 3

```python
p3=PolynomialFeatures(degree=3)

Xp3=p3.fit_transform(X.to_numpy().reshape(-1,1))
Xp3
```

```python
lr3=LinearRegression()
pr3=lr3.fit(Xp3,y)
```

```python
# Plotting

plt.figure(figsize=(10,8))
plt.scatter(X,y)
plt.plot(X,lr.predict(X.to_numpy().reshape(-1,1)),label='Lin Reg')
plt.plot(X,pr2.predict(Xp2),label='Poly Reg degree 2')

plt.plot(X,pr3.predict(Xp3),label='Poly Reg degree 3')

plt.legend();
```

# Polynomial degree 5

```python
p5=PolynomialFeatures(degree=5)
Xp5=p5.fit_transform(X.to_numpy().reshape(-1,1))
lr5=LinearRegression()
pr5=lr5.fit(Xp5,y)
```

```python
# Plotting

plt.figure(figsize=(10,8))
plt.scatter(X,y)
plt.plot(X,lr.predict(X.to_numpy().reshape(-1,1)),label='Lin Reg')
plt.plot(X,pr2.predict(Xp2),label='Poly Reg degree 2')
plt.plot(X,pr3.predict(Xp3),label='Poly Reg degree 3')

plt.plot(X,pr5.predict(Xp5),label='Poly Reg degree 5')


plt.legend();
```
