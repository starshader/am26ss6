## Imbalanced dataset

## Resolving imbalancing

### Upsampling 

### Downsampling

## Splitting the data

```python
bank_no=bank[bank['y']=='no']
bank_no.shape
```

```python
bank_no.head(25)
```

```python
bank_yes=bank[bank['y']=='yes']
bank_yes.shape
```

```python
bank_yes.head(25)
```

## Upsampling

```python
from sklearn.utils import resample

bank_yes_up=resample(bank_yes,n_samples=15000,replace=True,random_state=100)

bank_yes_up.head(25)
```

```python
bank_yes_up.shape
```

## Downsampling

On majority dataset

It reduces the contribution

```python
bank_no_down=resample(bank_no,replace=False,n_samples=25000,random_state=100)
bank_no_down.shape
```

```python
bank_no_down.head(25)
```

## Creating a new dataset

```python
bank_new=pd.concat([bank_no_down,bank_yes_up])
bank_new.shape
```

```python
bank_new.head(25)
```

```python
bank_new.tail(25)
```

## Shuffling the data


```python
from sklearn.utils import shuffle

bank_new=shuffle(bank_new)
bank_new.head(25)
```

## Separating to target and features

```python
# Target
y=bank_new['y']
y
```

```python
# Features

X=bank_new.drop(['y'],axis=1)
X
```

## Converting categorical features to numeric

```python
X_new=pd.get_dummies(X)
X_new
```

## Standardisation of features

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()

X_scaled=scaler.fit_transform(X_new)
X_scaled
```

## Splitting to train and test

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X_scaled,y,test_size=0.2,random_state=100)

X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

## Model building - K Nearest Neighbor

```python
from sklearn.neighbors import KNeighborsClassifier

knn=KNeighborsClassifier()

knn.fit(X_train,y_train)
```

## Performance analysis

```python
from sklearn.metrics import confusion_matrix, classification_report

report=classification_report(y_test, knn.predict(X_test))
cm=confusion_matrix(y_test, knn.predict(X_test))

print(' The CM:\n', cm)
print(' The report:\n', report)
```

## Hyper parameter tuning using GridSearchCV

```python
from sklearn.model_selection import GridSearchCV

knn_gs=GridSearchCV(knn, {'n_neighbors': range(2,8)})
```

```python
knn_gs.fit(X_train,y_train)
```

```python
knn_gs.best_params_
```

## Building the best model


```python
knn_best=KNeighborsClassifier(n_neighbors=2)
knn_best.fit(X_train,y_train)
```

```python
report=classification_report(y_test,knn_best.predict(X_test))

print(' Report:\n', report)
```
