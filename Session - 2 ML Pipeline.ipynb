### Collect the data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
sal=pd.read_csv('SalaryData.csv')
```

### Preprocess the data



y = b0 + b1. X

y = b0. 1 + b1. X

y= [b0 b1] [1 X] ^T

```python
import statsmodels.api as sm
```

```python
y=sal['Salary']
y
```

```python
X=sm.add_constant(sal['YearsExperience'])
X
```

### Divide the data into training and testing

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(
                 X,y,test_size=0.2, random_state=10)
```

### Perform Data Exploration

```python
plt.scatter(X_train['YearsExperience'], y_train, c='green')

plt.scatter(X_test['YearsExperience'], y_test, c='red')

plt.show();
```

### Model Building

#### Linear Regression - Method of Ordinary Least Squares



Functional form of Linear regression:

$ y= \beta{0} + \beta{1}.X + \epsilon$ 

```python
sal_lr=sm.OLS(y_train,X_train)
```

```python
# Training the model

sal_lr=sal_lr.fit()
```

```python
sal_lr
```

```python
# Value of parameters

sal_lr.params
```

### Model Diagnosis

```python
sal_lr.summary2()
```

### Residual Analysis

#### Assumption of regression




#### Checking for normal distribution of residuals

```python
# Calculating the residue
sal_lr.resid
```

```python
# Using probability - probability plot

# Compares the cumulative distribution of residuals with that of normal distribution.

probplot=sm.ProbPlot(sal_lr.resid)
probplot.ppplot(line='45')
plt.show();
```

### Remark

It shows that residuals are not very much normally disributed.

In the following figure, the residual are normally distributed.


### Test of homoscedasticity

It means that residuals have constant variance.

If the data points in residual plot do not have any pattern, it is satisfied.

It can be deduced from the residual plot.

```python
def standardisation(vals):
    return (vals-vals.mean()/vals.std())
```

```python
# Plotting residual plot


plt.scatter(standardisation(sal_lr.fittedvalues), 
            standardisation(sal_lr.resid))
plt.show();
```

### Remarks:


It shows that residuals have constant variance.

### Outlier detection

Many ways:

* Z- Score

* Mahalanobis distance

* Cook's Distance

* Leverage Value

#### Z- Score

It is the standradised distance of an observation from its mean.

Any observation with an Z-score of more than 3 is considered as an outlier.

```python
from scipy.stats import zscore

sal['zscore']=zscore(sal['Salary'])
```

```python
sal
```

```python
sal[sal['zscore']>3.0] 
```

```python
sal[sal['zscore']< -3.0]
```

### Remark:

There are no observations with z score more than 3.

So, there are no outliers.

## Making prediction and measuring accuracy

```python
y_pred= sal_lr.predict(X_test)
y_pred
```

```python
# Finding accuracy

from sklearn.metrics import r2_score, mean_squared_error

print(r2_score(y_test, y_pred))
```

```python
# RMSE 

print(np.sqrt(mean_squared_error(y_test,y_pred)))
```
